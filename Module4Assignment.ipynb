# Ques-4 Create an ordered list of HTML tags. Each list item must include the tag name and some information about the tag?

# Ans-4

```python
from IPython.core.display import display, HTML

html_code = '''
<!DOCTYPE html>
<html lang="en">
  <head>
    <meta charset="UTF-8" />
    <meta name="viewport" content="width=device-width, initial-scale=1.0" />
    <title>Ordered List</title>
  </head>
  <body>
    <ol>
      <li> <b>html – </b> Root element of an HTML page</li>
      <li> <b>head – </b> Contains metadata like title and links</li>
      <li> <b>title – </b> Sets the title shown in the browser tab</li>
      <li> <b>body – </b> Contains all the visible content of the page</li>
      <li> <b>h1 to h6 – </b> Headings, from largest (h1) to smallest (h6)</li>
      <li> <b>p – </b> Paragraph of text</li>
      <li> <b>a –  </b>Creates a hyperlink</li>
      <li> <b>img – </b> Displays an image</li>
      <li> <b>ul – </b> Unordered list (bullets)</li>
      <li> <b>ol – </b> Ordered list (numbers)</li>
      <li> <b>li – </b> List item inside ul or ol</li>
      <li> <b>div –  </b>Block-level container for layout</li>
      <li> <b>span – </b> Inline container for styling</li>
      <li> <b>br – </b> Line break</li>
      <li> <b>input – </b> Input field for forms</li>
    </ol>
  </body>
</html>
'''
print("HTML Code:")
print(html_code)

print("\nRendered Output:")
display(HTML(html_code))
```

# Ques-8) Create a table to display a conference schedule. Each row corresponds to a time slot, and each column corresponds to a room. Some time slots might have multiple sessions running simultaneously in different rooms. Utilize rowspan and colspan attributes as necessary to accommodate this complex schedule. (use table attribute “cellpadding”  to give extra padding in each table cell )?

```python
from IPython.core.display import display, HTML

html_code = '''
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Document</title>

    <style>
        .main{
            border: 3px solid black;
            display: inline-block;
            padding: 22px;
        }
        table{
            border-collapse: collapse;
        }
    </style>
</head>
<body>
    <main class="main">
       <h1>Conference Schedule</h1>

       <table border="1" cellpadding="10">
          <tr>
            <th>Time</th>
            <th>Room 1</th>
            <th>Room 2</th>
            <th>Room 3</th>
            <th>Room 4</th>
          </tr>
          <tr>
            <td rowspan="3">9:00 AM - 10:00 AM</td>
            <td rowspan="2">Keynote</td>
            <td>Session A</td>
            <td>Session B</td>
            <td rowspan="3">Session C</td>
          </tr>
          <tr>
            <td>Session D</td>
            <td>Session E</td>
          </tr>
          <tr>
            <td>10:30 AM - 11:30 AM</td>
            <td colspan="2">Session F</td>
          </tr>

          <tr>
            <td>12:00 PM - 1:00 PM</td>
            <td colspan="4">Lunch Break</td>
          </tr>
          <tr>
            <td rowspan="2">1:00 PM - 2:00 PM</td>
            <td>Session G</td>
            <td rowspan="2">Session H</td>
            <td>Session I</td>
            <td>Session J</td>
          </tr>
          <tr>
            <td>Session K</td>
            <td>Session L</td>
            <td>Session M</td>
          </tr>
       </table>
    </main>
</body>
</html>
'''

print("HTML Code:")
print(html_code)

print("\nRendered Output:")
display(HTML(html_code))
```
